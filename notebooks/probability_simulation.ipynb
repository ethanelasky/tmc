```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
K = 100 # Number of topics
N = 48000 # Number of documents, total (to change when finished scraping)
```

```python
p = [1 / K for i in range(K)]
ns = np.arange(10, N)
cover_probs = np.zeros_like(ns).astype(float)

for i in range(len(ns)):
    n = ns[i]
    X = stats.multinomial.rvs(n, p, size=1000)
    cover_probs[i] = np.mean(np.all(X > 0, axis=0))
```

```python
plt.plot(ns, cover_probs)
plt.ylim(0, 1.1)
plt.xlim(1000, 1250)

np.mean(cover_probs[1200:1300])
```
