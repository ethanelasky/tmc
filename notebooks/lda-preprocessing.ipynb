```python
stopwords = read_stopwords_list()

# Load the Chinese language model
nlp = zh_core_web_sm.load(disable=['ner', 'parser', 'attribute_ruler', 'tagger', 'tok2vec'])
def nlpize(doc):
    doc = nlp(doc) 
    return [w.text for w in doc if w.text not in stopwords and w.text not in baidu_stopwords]
```

```python
seed_words = ["　", ".", "\n", "-", "／", "／", '中國','大陸','美國', '中共', '北京主權', '中華人民共和國']

def remove_stopwords(lst):
    return [i.strip() for i in lst if i.strip() and i not in seed_words]
```

```python
df = pd.read_parquet("china_us_contents.parquet")
```

```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# Assuming you have a DataFrame named 'df'
chunk_size = 10000

# Calculate the number of chunks
num_rows = len(df)
num_chunks = num_rows // chunk_size + 1

# Iterate over the chunks and save each chunk to a separate Parquet file
for i in range(0, num_chunks):
    start_index = i * chunk_size
    end_index = min((i + 1) * chunk_size, num_rows)
    
    chunk_df = df.iloc[start_index:end_index]

    chunk_df = chunk_df['text'].astype(str).apply(nlpize).apply(remove_stopwords)
    chunk_df.to_frame().to_parquet(f'chunks/processed/china_us_tokenized_{i}.parquet')
```
